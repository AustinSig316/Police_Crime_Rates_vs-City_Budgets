```python
import pandas as pd
import numpy as np
```

```python
alabama_poverty_df = pd.read_csv('Resources/PovertyReport Alabama.csv')
```

```python
alabama_poverty_df.head()
```

```python
new_alabama_poverty_df = alabama_poverty_df.loc[0:67]
new_alabama_poverty_df.head()
```

```python
new_alabama_poverty_df = new_alabama_poverty_df[['county2', 'total_est_pct3']]
new_alabama_poverty_df = new_alabama_poverty_df.rename(columns={'county2': "county", 'total_est_pct3': 'poverty_pct'})

new_alabama_poverty_df['state'] = 'Alabama'

new_alabama_poverty_df = new_alabama_poverty_df[['county', 'state', 'poverty_pct']]

new_alabama_poverty_df.drop(index=new_alabama_poverty_df.index[0], 
        axis=0, 
        inplace=True)
new_alabama_poverty_df
```
